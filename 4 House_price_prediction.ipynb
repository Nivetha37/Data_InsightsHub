```python
data.drop(columns=['price_per_sqft','hasYard','hasPool','cityCode','cityPartRange','made','hasStormProtector','basement','attic','garage','hasStorageRoom','hasGuestRoom'],inplace=True)
data.head()
```

```python
x=data.drop(columns=['price'])
y=data['price']
```

```python
#Training and testing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_selector
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state =0)
```

```python
print(X_train.shape)
print(X_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)

lr = LinearRegression()
lr.fit(X_scaled, y)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse=False), ['numPrevOwners']), remainder='passthrough')
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(Y_test,y_pred_lr)
```

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(Y_test,y_pred_lasso)
```

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(Y_test,y_pred_ridge)
```

```python
print("No Regularization: ", r2_score(Y_test,y_pred_lr))
print("Lasso: ", r2_score(Y_test,y_pred_lasso))
print("Ridge: ", r2_score(Y_test,y_pred_ridge))
```
